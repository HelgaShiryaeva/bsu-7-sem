```python
% matplotlib inline

import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.utils import shuffle
from scipy.spatial import distance_matrix
import numpy as np
from sklearn.metrics import precision_recall_fscore_support
```

```python
wine = datasets.load_wine()
X = wine.data
Y = wine.target
target_names = wine.target_names
L = len(target_names)

lda = LinearDiscriminantAnalysis(n_components=2)
X = lda.fit(X, Y).transform(X)

#pca = PCA(n_components=2)
#X = pca.fit(X).transform(X)

X, Y = shuffle(X, Y, random_state=0)
```

```python
def plot_samples(X, Y, target_names, title='Samples'):
    plt.figure()
    colors = ['navy', 'turquoise', 'darkorange']
    lw = 2

    for color, i, target_name in zip(colors, [0, 1, 2], target_names):
        plt.scatter(X[Y == i, 0], X[Y == i, 1], color=color, alpha=.8, lw=lw,
                    label=target_name)
    plt.legend(loc='best', shadow=False, scatterpoints=1)
    plt.title(title)
    plt.show()

plot_samples(X, Y, target_names, 'LDA of Wine dataset')
```

## Training stage

```python
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x), axis=0)

def prediction_metric(x):
    return 1 / x.mean()


def run_and_evaluate(xt, yt, xp, p, L, metric):
    n = len(xt)
    D = distance_matrix(xt, xp, p=p )
    B = np.zeros((n, L))

    for row in range(n):
        for cls in range(L):
            B[row][cls] = metric(D[row][yp == cls])

    B = np.array([softmax(b) for b in B])
    predicted_classes = B.argmax(axis=1)

    #print(B[:5])
    #print(yt[:5])

    accuracy = (yt == predicted_classes).mean()
    result = precision_recall_fscore_support(yt, predicted_classes)

    print('\nPrecision')
    print(result[0])
    print('\nRecall')
    print(result[1])
    print('\n Accuracy: {}%'.format(accuracy * 100))
    
    return predicted_classes
    
predicted_classes = run_and_evaluate(xt, yt, xp, 2, L, prediction_metric)
    
predicted_classes = run_and_evaluate(xt, yt, xp, 1, L, prediction_metric)
```
