### Осуществить моделирование n = 1000 реализаций СВ из заданных дискретных распределений. 

```python
p_bernoulli = 0.9
p_geometric = 0.7
n = 1000
```

```python
from random import random

def bernoulli(p, n):
    for i in range (n):
        yield 1 if p < random() else 0
```

```python
bernoulli_seq = list(bernoulli(p_bernoulli, n))

for value in bernoulli_seq:
    print(value)
```

```python
def get_bernoulli_frequencies(seq):
    zeroes = 0
    ones = 0
    
    for value in seq:
        if value is 0:
            zeroes += 1
        else:
            ones += 1
    
    return [zeroes, ones]
```

```python
freqs = get_bernoulli_frequencies(bernoulli_seq)

print(freqs)
```

```python
def geometric(p, n):
    for i in range(n):
        b = list(bernoulli(p, int(n / 10)))
        i = 0
        while i < len(b):
            if b[i] == 1:
                break
            i += 1
        yield i
    
```

```python
geometric_seq = list(geometric(p_geometric, n))

for value in geometric_seq:
    print(value)
```

```python
def expectation(seq):
    sum = 0.
    for value in seq:
        sum += value
        
    return sum / len(seq)
```
