### Осуществить моделирование n = 1000 реализаций СВ из заданных дискретных распределений. 

```python
p_bernoulli = 0.9
p_geometric = 0.7
n = 10000
```

```python
from random import random

def bernoulli(p, n):
    for i in range (n):
        yield 0 if p < random() else 1
```

```python
bernoulli_seq = list(bernoulli(p_bernoulli, n))

for value in bernoulli_seq:
    print(value)
```

```python
def get_bernoulli_frequencies(seq):
    zeroes = 0
    ones = 0
    
    for value in seq:
        if value is 0:
            zeroes += 1
        else:
            ones += 1
    
    return [zeroes, ones]
```

```python
freqs = get_bernoulli_frequencies(bernoulli_seq)

print(freqs)
```

```python
def geometric(p, n):
    for i in range(n):
        b = list(bernoulli(p, int(n)))
        i = 0
        while i < len(b):
            if b[i] == 1:
                break
            i += 1
        yield i
    
```

```python
geometric_seq = list(geometric(p_geometric, n))

for value in geometric_seq:
    print(value)
```

```python
def expectation(seq):
    sum = 0.
    for value in seq:
        sum += value
        
    return sum / len(seq)
```

```python
bernoulli_expectation = p_bernoulli
bernoulli_variance = p_bernoulli * (1 - p_bernoulli)
geometric_expectation = 1 / p_geometric
geometric_variance = (1 - p_geometric) / p_geometric ** 2
```

```python
def variance(seq, exp):
    var = 0.
    for value in seq:
        var += (value - exp) ** 2
    
    return var / len(seq)
```

```python
bernoulli_exp = expectation(bernoulli_seq)
bernoulli_var = variance(bernoulli_seq, bernoulli_exp)

geometric_exp = expectation(geometric_seq)
geometric_var = variance(geometric_seq, geometric_exp)
```

```python
print('Bernoulli')
print('Theoretic : E = {}, D = {}'.format(bernoulli_expectation, bernoulli_variance))
print('Actual :    E = {}, D = {}'.format(bernoulli_exp, bernoulli_var))
print()
print('Geometric')
print('Theoretic : E = {}, D = {}'.format(geometric_expectation, geometric_variance))
print('Actual :    E = {}, D = {}'.format(geometric_exp, geometric_var))
```
